```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
m = 1700
```

```python
x = np.linspace(-1025, 975, 20)
y = np.linspace(-1050, 250, 20)
xx, yy = np.meshgrid(x, y)
```

```python
x1, y1 = -455, 0
x2, y2 = 455, 0
```

```python
u = m * (xx - x1) / ((xx - x1)**2 + (yy - y1)**2) + \
    m * (xx - x2) / ((xx - x2)**2 + (yy - y2)**2)
v = m * (yy - y1) / ((xx - x1)**2 + (yy - y1)**2) + \
    m * (yy - y2) / ((xx - x2)**2 + (yy - y2)**2)
```

```python
for j in range(xx.shape[0]):
    for i in range(xx.shape[1]):
        if np.sqrt((xx[j, i] - 455)**2 + yy[j, i]**2) < 150:
            u[j, i] = np.nan
            v[j, i] = np.nan

        elif np.sqrt((xx[j, i] + 455)**2 + yy[j, i]**2) < 150:
            u[j, i] = np.nan
            v[j, i] = np.nan
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
plt.xlim(-1025, 975)
plt.ylim(-1050, 250)

plt.quiver(xx, yy, u, v, np.sqrt(u**2 + v**2), cmap="jet")

plt.colorbar()
plt.clim(0, 10)
plt.show()
```
