```python
ptv_result = [0] * 10
```

```python
columns = ["x", "y", "dx", "dy", "p0", "p1"]
```

```python
for i in tqdm(range(5100, 5200, 10)):
    try:
        r = pd.read_csv("ptv/" + f"pp_{i}.csv", delimiter=',')
        r = pd.DataFrame(data=r.values, columns=columns)
    except StopIteration:
        r = None
    except OSError:
        r = None

    if r is None:
        pass
    else:
        for row in r.index:
            if (r.loc[row] == 0).all():
                r.drop(row, axis=0, inplace=True)
        ptv_result[(i-5100)//10] = r
```

```python
result = np.zeros([10, 1000, 6])  # x, y, dx, dy, p1, flag
result[:, :, :-1] = np.nan

# 5000
tmp = ptv_result[0].sort_values(by='p0', ascending=True)
p_max = tmp.values.shape[0]
```

```python
tmp
```

```python
result[0, :p_max] = np.hstack([np.hstack([tmp.values[:, :4], tmp.values[:, 5:]]), np.ones([p_max, 1])])
```

```python
# 5010
tmp = ptv_result[1].sort_values(by='p0', ascending=True)  # 10
p = p_max

for j in range(tmp.values.shape[0]):
    is_exist = False

    for i in range(p_max):
        if is_exist:
            continue

        if tmp.values[j, 4] == result[0, i, 4]:
            result[1, i, :4] = tmp.values[j, :4]
            result[1, i, 4] = tmp.values[j, 5]
            result[1, i, 5] = 1
            is_exist = True

    if not is_exist:
        result[1, p, :4] = tmp.values[j, :4]
        result[1, p, 4] = tmp.values[j, 5]
        result[1, p, 5] = 1
        p += 1

for i in range(p_max):
    if result[1, i, 5] == 0:
        result[1, i, 0] = result[0, i, 0] + result[0, i, 2]
        result[1, i, 1] = result[0, i, 1] + result[0, i, 3]
        result[1, i, 2:4] = result[0, i, 2:4]
        result[1, i, 4] = -1  # predict

p_max = p
```

```python
# 5020
tmp = ptv_result[2].sort_values(by='p0', ascending=True)  # 20
p = p_max

for j in range(tmp.values.shape[0]):
    is_exist = False

    for i in range(p_max):
        if is_exist:
            continue

        if tmp.values[j, 4] == result[1, i, 4]:
            result[2, i, :4] = tmp.values[j, :4]
            result[2, i, 4] = tmp.values[j, 5]
            result[2, i, 5] = 1
            is_exist = True

        elif result[1, i, 4] == -1 and \
            np.sqrt(
                (result[1, i, 0] + result[1, i, 2] - tmp.values[j, 0]) ** 2 + \
                (result[1, i, 1] + result[1, i, 3] - tmp.values[j, 1]) ** 2
            ) < 5:
            result[2, i, :4] = tmp.values[j, :4]
            result[2, i, 4] = tmp.values[j, 5]
            result[2, i, 5] = 1
            is_exist = True

    if not is_exist:
        result[2, p, :4] = tmp.values[j, :4]
        result[2, p, 4] = tmp.values[j, 5]
        result[2, p, 5] = 1
        p += 1

for i in range(p_max):
    if result[2, i, 5] == 0:
        result[2, i, 0] = result[1, i, 0] + result[1, i, 2]
        result[2, i, 1] = result[1, i, 1] + result[1, i, 3]
        result[2, i, 2:4] = result[1, i, 2:4]
        result[2, i, 4] = -1  # predict

p_max = p
```
