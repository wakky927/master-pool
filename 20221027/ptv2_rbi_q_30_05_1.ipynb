```python
import pickle

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
import warnings
warnings.simplefilter("ignore", UserWarning)
```

```python
IN_DIR = "D:\\result\\20221027\\"
OUT_DIR = "D:\\result\\20221027\\"
sub_dir_list = "rbi_q_30_05\\1\\"
```

```python
time_table = pd.read_csv("time.csv", header=None, index_col=0)
time_table = time_table.values[9, :43].astype(np.int)
```

```python
ptv_result = np.zeros((1, 4))
```

```python
for t in range(1, 11):
    for i in tqdm(range(t, 10900, 10)):
        for j in range(21):
            if time_table[2 * j] <= i < time_table[2 * j + 1] or time_table[42] <= i:
                try:
                    r = np.loadtxt(OUT_DIR + "ptv\\" + sub_dir_list + f"pp_{i}.csv", delimiter=',', skiprows=1)
                except StopIteration:
                    r = None
                except OSError:
                    r = None

                if r is None:
                    pass
                elif r.ndim == 1:
                    if len(r) > 0:
                        r = r[np.newaxis, :]
                        ptv_result = np.vstack([ptv_result, r])

                elif r.ndim == 2:
                    ptv_result = np.vstack([ptv_result, r])
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
plt.xlim(0, 1024)
plt.ylim(0, 600)
c = ax.quiver(ptv_result[1:, 0], 500-ptv_result[1:, 1], ptv_result[1:, 2], -ptv_result[1:, 3], np.sqrt(ptv_result[1:, 2]**2 + ptv_result[1:, 3]**2), cmap="jet")
fig.colorbar(c)
plt.show()
```

```python
with open(f'ptv_rbi_q_30_05_1.pickle', mode='wb') as fp:
    pickle.dump(ptv_result, fp)
```
